Récupération des joueurs d'un club :

    - Faire une requête HTTP
    - Identifier la structure de la page
    - Extraire les données des joueurs

```python
import requests
from bs4 import BeautifulSoup
import pandas as pd

# Page de club du PSG
url = "https://www.transfermarkt.fr/paris-saint-germain/kader/verein/583"

headers = {
    "User-Agent": "Mozilla/5.0"
}

response = requests.get(url, headers=headers)
soup = BeautifulSoup(response.content, "html.parser")
```

```python
players_table = soup.find("table", {"class": "items"})
players = []

for row in players_table.find_all("tr", class_=["odd", "even"]):
    name_cell = row.find("td", class_="hauptlink")
    if name_cell:
        name = name_cell.text.strip()
        age = row.find_all("td")[5].text.strip()
        position = row.find_all("td")[4].text.strip()
        market_value = row.find("td", class_="rechts").text.strip()
        numero = row.find("div", class_="rn_nummer").text.strip()
        nationnalities = row.find_all("td")[6]
        natList = [img["alt"] for img in nationnalities.find_all("img")]

        players.append({
            "Nom": name,
            "Âge": age,
            "Position": position,
            "Valeur marchande": market_value,
            "Numéro": numero,
            "nationnalite": natList
        })

df = pd.DataFrame(players)
df
```
